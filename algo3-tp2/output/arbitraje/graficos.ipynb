```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
bellmanTest1 = pd.read_csv('test1-bellman-ford.csv')
floydTest1 = pd.read_csv('test1-floyd.csv')

bellmanTest2SinArbitraje = pd.read_csv('test2-bellman-ford-sin-arbitraje.csv')
floydTest2SinArbitraje = pd.read_csv('test2-floyd-sin-arbitraje.csv')

bellmanTest2ConArbitraje = pd.read_csv('test2-bellman-ford-con-arbitraje.csv')
floydTest2ConArbitraje = pd.read_csv('test2-floyd-con-arbitraje.csv')

bellmanTest2ConArbitrajeHamiltoneano = pd.read_csv('test2-bellman-ford-con-arbitraje-hamiltoneano.csv')
floydTest2ConArbitrajeHamiltoneano = pd.read_csv('test2-floyd-con-arbitraje-hamiltoneano.csv')
```

```python
#TEST 1 Complejidades Arbitraje
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

bellmanTest1['FB/N^3'] = bellmanTest1['TIEMPO'] / bellmanTest1['N']**3;
ax1 = sns.lineplot(x="N", y="FB/N^3", data=bellmanTest1.reset_index(), label="Bellman-Ford", ax=ax);
floydTest1['FB/N^3'] = floydTest1['TIEMPO'] / floydTest1['N']**3;
ax1 = sns.lineplot(x="N", y="FB/N^3", data=floydTest1.reset_index(), label="Floyd", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

plt.ylim(-0.01,0.01);

plt.ylabel(r"F(N,V) = $\frac{f(N, V)}{N^3}$", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

```python
#TEST 2 Comparacion instancias sin arbitraje siempre
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)


ax1 = sns.lineplot(x="N", y="TIEMPO", data=bellmanTest2SinArbitraje.reset_index(), label="Bellman-Ford",ax=ax);
ax1 = sns.lineplot(x="N", y="TIEMPO", data=floydTest2SinArbitraje.reset_index(), label="Floyd",ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

plt.ylabel("F(N,V) [ms]", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14);
plt.setp(ax.lines,linewidth=3);
```

```python
#TEST 2 Comparacion instancias con arbitraje siempre
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)


ax1 = sns.lineplot(x="N", y="TIEMPO", data=bellmanTest2ConArbitraje.reset_index(), label="Bellman-Ford",ax=ax);
ax1 = sns.lineplot(x="N", y="TIEMPO", data=floydTest2ConArbitraje.reset_index(), label="Floyd",ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

plt.ylabel("F(N,V) [ms]", fontsize=16);
plt.xlabel("N", fontsize=16);
#plt.ylim(0,12000)
plt.tick_params(labelsize=14);
plt.setp(ax.lines,linewidth=3);
```
