```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
savingsTiemposA = pd.read_csv('1-savings/A/tiempo.csv')
savingsOptimosA = pd.read_csv('1-savings/A/optimo.csv')
savings = savingsTiemposA.merge(savingsOptimosA, on='n')

golosaTiemposA = pd.read_csv('2-golosa/A/tiempo.csv')
golosaOptimosA = pd.read_csv('2-golosa/A/optimo.csv')
golosa = golosaTiemposA.merge(golosaOptimosA, on='n')

sweepAlgorithmTiemposA = pd.read_csv('3-sweep-algorithm/A/tiempo.csv')
sweepAlgorithmOptimosA = pd.read_csv('3-sweep-algorithm/A/optimo.csv')
sweepAlgorithm = sweepAlgorithmTiemposA.merge(sweepAlgorithmOptimosA, on='n')

otraClusterizacionTiemposA = pd.read_csv('4-otra-clusterizacion/A/tiempo.csv')
otraClusterizacionOptimosA = pd.read_csv('4-otra-clusterizacion/A/optimo.csv')
otraClusterizacion = otraClusterizacionTiemposA.merge(otraClusterizacionOptimosA, on='n')

simulatedAnnealingTiemposA = pd.read_csv('5-simulated-annealing/A/tiempo.csv')
simulatedAnnealingOptimosA = pd.read_csv('5-simulated-annealing/A/optimo.csv')
simulatedAnnealing = simulatedAnnealingTiemposA.merge(simulatedAnnealingOptimosA, on='n')
```

# SAVINGS

```python
#SAVINGS Complejidad savings (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

savings['SAV/N^3'] = savings['tiempo'] / savings['n']**3;
ax1 = sns.lineplot(x="n", y="SAV/N^3", data=savings.reset_index(), ax=ax);

#Reverse legends
#handles, labels = ax1.get_legend_handles_labels()
#ax1.legend(reversed(handles), reversed(labels), title='Líneas', loc='upper right')

plt.ylim(-0.02,0.02);
sns.set_style("whitegrid")
plt.ylabel(r"Relación $\frac{Savings(G)}{n^3}$", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```

```python
#SAVINGS Resultado vs optimo (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

ax1 = sns.lineplot(x="n", y="resultado", data=savings.reset_index(), label="Saving", ax=ax);
ax2 = sns.lineplot(x="n", y="optimo", data=savings.reset_index(), label="Optimo", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='Líneas', loc='upper right')

#plt.ylim(-0.01,0.01);

plt.ylabel(r"Distancia recorrida", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```

```python



#SAVINGS Bar plot Diferencias entre Resultado vs optimo (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
savings['distanciaOptRes'] = savings['resultado'] / savings['optimo']
ax1 = sns.barplot(x="n", y="distanciaOptRes", data=savings,ax=ax,palette=["#006faf"])

plt.ylabel(r"Relación $\frac{distancia(Savings)}{distancia(óptimo)}$", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```

# SIMULATED-ANNEALING

```python
#SIMULATED-ANNEALING Complejidad SIMULATED-ANNEALING (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
R=3

distanciaCienPorCien =  282.8427

simulatedAnnealing['SIM/N^3'] = simulatedAnnealing['tiempo'] / ((simulatedAnnealing['n'])**3 * R * np.log(distanciaCienPorCien * simulatedAnnealing['n'] ) );
ax1 = sns.lineplot(x="n", y="SIM/N^3", data=simulatedAnnealing.reset_index(), ax=ax);

plt.ylim(-0.02,0.02);
plt.ylabel(r"Relación $\frac{SimulatedAnnealing(G)}{n^3 \times R \times C_i}}$", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```

```python
#SIMULATED-ANNEALING Resultado vs optimo (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

ax1 = sns.lineplot(x="n", y="resultado", data=simulatedAnnealing.reset_index(), label="Simulated-Annealing", ax=ax);
ax2 = sns.lineplot(x="n", y="optimo", data=simulatedAnnealing.reset_index(), label="Optimo", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='Líneas', loc='upper right')

#plt.ylim(-0.01,0.01);

plt.ylabel(r"Distancia recorrida", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```
